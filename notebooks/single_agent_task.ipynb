# Single agent task

### **RQ**: can an embodied agent manipulating a substrate learn to output the correct numerosity of a scene?

A single agent is presented with scenes containing objects of different numerosities. 
The agent has two fingers, and it can manipulate a grid-like substrate which it can use as aid in the task.
It can move one finger over the scene and the other one over the grid-like substrate.
In each episode, it can perform all these actions and, finally, output a label (1-hot or binary) which should represent the numerosity of elements in the scene.
The agent is trained using a RL algorithm (e.g. Q-Learning). It is rewarded when the final label is correct.
